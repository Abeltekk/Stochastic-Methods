```python
from pylab import *
import numpy as np
from scipy import special
import matplotlib.pyplot as plt


def rel_error(n):
    stirling= np.sqrt(2*np.pi*n)*(n/np.e)**n
    actual_val= special.factorial(n)
    return (actual_val- stirling)/stirling



n= arange(1,100)
plt.loglog(n, rel_error(n))

plt.xlabel('n')
plt.ylabel('Relative Error')
show()

#the slope is very close to -1
#the next order of the stirling approximation should be to f(n)/12*n
```
